```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 读取图像
image = cv2.imread('book.jpg')

image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)

# 将图像转换为灰度图像
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 使用高斯模糊减少噪声
imgBlur = cv2.GaussianBlur(gray_image, (3, 5), 0)

all = []
name = []
# 进行Canny边缘检测
imgCanny = cv2.Canny(imgBlur, 10, 45)
all.append(imgCanny)
name.append("Canny")

plt.figure(figsize=(10, 5))
# 显示原始图像
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 显示Canny边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(imgCanny, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

```python
# 定义Robert算子的核
robert_x = np.array([[1, 0],
                     [0, -1]])

# 应用Robert算子
edges = cv2.filter2D(imgBlur, -1, robert_x)
all.append(edges)
name.append("Robert")

plt.figure(figsize=(10, 5))
# 显示原始图像
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 显示Cobert算子边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Robert Edge Detection')
plt.axis('off')
plt.show()
```

```python
# 定义Prewitt算子的核
prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])

prewitt_y = np.array([[-1, -1, -1],
                      [0, 0, 0],
                      [1, 1, 1]])

# 应用Prewitt算子
edges_x = cv2.filter2D(imgBlur, -1, prewitt_x)
edges_y = cv2.filter2D(imgBlur, -1, prewitt_y)

# 计算梯度幅值
edges = np.sqrt(np.square(edges_x) + np.square(edges_y))
all.append(edges)
name.append("Prewitt")

plt.figure(figsize=(10, 5))
# 显示原始图像
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 显示Prewitt算子边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Prewitt Edge Detection')
plt.axis('off')
plt.show()
```

```python
# 使用Sobel算子进行边缘检测
sobel_x = cv2.Sobel(imgBlur, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(imgBlur, cv2.CV_64F, 0, 1, ksize=3)

# 计算梯度幅值
edges = np.sqrt(np.square(sobel_x) + np.square(sobel_y))
all.append(edges)
name.append("Sobel")

plt.figure(figsize=(10, 5))
# 显示原始图像
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 显示Prewitt算子边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')
plt.show()
```

```python
# 定义Robinson算子的卷积核
robinson_x = np.array([[1, 2, 1],
                        [0, 0, 0],
                        [-1, -2, -1]])

robinson_y = np.array([[1, 0, -1],
                        [2, 0, -2],
                        [1, 0, -1]])

# 使用Robinson算子进行边缘检测
edges_x = cv2.filter2D(imgBlur, -1, robinson_x)
edges_y = cv2.filter2D(imgBlur, -1, robinson_y)

# 计算梯度幅值
edges = np.sqrt(np.square(edges_x) + np.square(edges_y))
all.append(edges)
name.append("Robinson")

plt.figure(figsize=(10, 5))
# 显示原始图像
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 显示Prewitt算子边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Robinson Edge Detection')
plt.axis('off')
plt.show()
```

```python
# 使用Laplacian算子进行边缘检测
edges = cv2.Laplacian(imgBlur, cv2.CV_64F)

# 将边缘结果转换为绝对值
edges = np.uint8(np.absolute(edges))
all.append(edges)
name.append("Laplacian")

plt.figure(figsize=(10, 5))
# 显示原始图像
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 显示Prewitt算子边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')
plt.show()
```

```python
num_images = len(all)
plt.figure(figsize=(15, 5))
for i in range(num_images):
    plt.subplot(1, 6, i+1)
    plt.imshow(all[i], cmap='gray')
    plt.title(name[i])
    plt.axis('off')

plt.show()
```

1）Robert算子对陡峭的低噪声图像效果较好，尤其是边缘正负45度较多的图像，但定位准确率较差；

2）Prewitt算子对灰度渐变的图像边缘提取效果较好，而没有考虑相邻点的距离远近对当前像素点的影响；

3）Sobel算子考虑了综合因素，对噪声较多的图像处理效果更好。

4）Laplacian算子对噪声比较敏感，由于其算法可能会出现双像素边界，常用来判断边缘像素位于图像的明区或暗区，很少用于边缘检测；
 
5）Canny:低错误率： 算法能够尽可能地减少边缘检测中的错误检测率，尤其是对于噪声较少的图像,准确的定位：能够精确地定位图像中的边缘，
即使边缘是模糊的或者存在一定程度的噪声。可以检测到图像中的多条边缘，并且能够在边缘之间产生一定的间隔，使得边缘检测结果更加清晰。

6）Robinson算子多方向检测,可以在八个不同方向上进行边缘检测，而不仅仅是水平和垂直方向。更强的边缘响应：在检测图像中的边缘时，
通常能够产生更强的响应，这使得它在一些情况下可能会比 Sobel 算子更加敏感。简单性：相对于其他更复杂的算法来说，实现起来比较简单，而且
计算速度较快。

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 读取图像
image = cv2.imread('book.jpg')

image = cv2.resize(image, (0, 0), fx=0.41, fy=0.41)

# 将图像转换为灰度图像
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 使用高斯模糊减少噪声
imgBlur = cv2.GaussianBlur(gray_image, (11, 11), 0)

# 进行Canny边缘检测
imgCanny = cv2.Canny(imgBlur, 11, 25)

# 创建一个5x5的卷积核
kernel = np.ones((5, 5))

imgDial = cv2.dilate(imgCanny, kernel, iterations=11)  # 膨胀
 
imgThre = cv2.erode(imgDial, kernel, iterations=10)  # 腐蚀

plt.figure(figsize=(5, 5))
plt.imshow(imgThre, cmap='gray')
plt.axis('off')

plt.show()
```

对书本进行高斯滤波，Canny算子检测边缘，再通过膨胀和腐蚀连接用来填充物体内的小空洞，连接断开的轮廓线。

```python
def detect_rectangle(image):
    
    # 寻找轮廓
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    #将数组扩充成3通道图
    image_copy = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    
    # 筛选出长方形轮廓
    rectangles = []
    for contour in contours:
        approx = cv2.approxPolyDP(contour, 0.02 * cv2.arcLength(contour, True), True)
        if len(approx) == 4:
            rectangles.append(approx)
            
    # 获取矩形轮廓的边缘点
    edge_points = []
    for rectangle in rectangles:
        for point in rectangle:
            edge_points.append(tuple(point[0]))
    print(rectangles)
    # 绘制长方形轮廓
    for rectangle in rectangles:
        cv2.drawContours(image_copy, [rectangle], -1, (0, 0, 255), 2)
    
    # 显示图像
    plt.imshow(cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
     
    return edge_points

edge_points = detect_rectangle(imgThre)
```

得到书本的四个边缘点,根据这些边缘点进行透视变换，得到图片矫正的效果

```python
def perspective_transform(image, src_points, dst_size):
    # 定义目标点
    dst_points = np.float32([[0, 0], [dst_size[0], 0], [0, dst_size[1]], [dst_size[0], dst_size[1]]])
    
    # 计算透视变换矩阵
    M = cv2.getPerspectiveTransform(np.float32(src_points), dst_points)
    
    # 进行透视变换
    warped_image = cv2.warpPerspective(image, M, dst_size)
    
    return warped_image

def sort_points(points):
    # 按照 x 坐标进行排序，找到左上和左下顶点
    points = sorted(points, key=lambda x: x[0])
    left_points = sorted(points[:2], key=lambda x: x[1])
    top_left, bottom_left = left_points
    
    # 按照 x 坐标进行排序，找到右上和右下顶点
    right_points = sorted(points[2:], key=lambda x: x[1])
    top_right, bottom_right = right_points
    
    return [top_left, top_right, bottom_left, bottom_right]

edge_points = sort_points(edge_points)
# 进行透视变换
warped_image = perspective_transform(image, edge_points, (500, 700))

# 显示图像
plt.imshow(cv2.cvtColor(warped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

利用霍夫变换进行圆检测，得到硬币和大圆的半径

```python
def detect_circle(image):

    #将数组扩充成3通道图
    image_copy = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    
    circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, 1, 50,
                               param1=20, param2=54, minRadius=10, maxRadius=100)
    print(circles)
    # 如果找到圆
    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        for (x, y, r) in circles:
            cv2.circle(image_copy, (x, y), r, (0, 255, 0), 2)
    # 显示图像
    plt.imshow(cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
    return circles

imgWarp = warped_image.copy()
imgGray = cv2.cvtColor(imgWarp, cv2.COLOR_RGB2GRAY)
# 中值滤波
imgblur = cv2.medianBlur(imgGray, 5)      
    
circles = detect_circle(imgblur)
```

通过比例计算答案

```python
height, width = imgblur.shape
coin = 2.5 / 2 / circles[1][2]
print("书本的长与宽:", height * coin, " and " ,width * coin)
print("书本右上方用铅笔画的圆圈的外圆直径:",2 * circles[0][2] * coin)
```
